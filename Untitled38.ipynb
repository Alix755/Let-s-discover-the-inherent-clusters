```python
Sum_of_squarred_distances=[]
K=range(1,15)
for k in K:
    km=KMeans(n_clusters=k)
    km=km.fit(df)
    Sum_of_squarred_distances.append(km.inertia_)
```

```python
plt.plot(K,Sum_of_squarred_distances,'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squarred_distances')
plt.title('Elbow Method For Optimal k')
plt.show
```

# Here, we can see that the elbow is situated in k=3, so we can conclude that the optimal k is 3
